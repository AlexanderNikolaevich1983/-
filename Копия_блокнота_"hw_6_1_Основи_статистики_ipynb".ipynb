Виведіть результат кожного обчислення на екран.

```python

x = [1, 3, 4]

sigma_x = sum(x)
sigma_x2 = sum(i**2 for i in x)
sigma_x_minus_1_sq = sum((i - 1)**2 for i in x)

print(f"Σx = {sigma_x}")
print(f"Σx^2 = {sigma_x2}")
print(f"Σ(x-1)^2 = {sigma_x_minus_1_sq}")
```

**Завдання 3.** Для того ж набору даних в змінній `sum_purchases` обчисліть **медіану** кожним з зазначених методів:  

  3.1. без використання жодних бібліотек, користуючись тільки Python.  
  3.2. використовуючи відповідний метод бібліотеки numpy.  

Виведіть результат обчислень в кожному з випадків на екран. Порівняйте результати зі значенням вибіркового середнього. Напишіть коментар, для яких задач ви б використовували вибіркове середнє, а для яких - медіану?

```python
# 3.1. Обчислення медіани без бібліотек
def median_manual(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid = n // 2
    if n % 2 == 0:
        return (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
        return sorted_data[mid]

median_manual_value = median_manual(sum_purchases)
print(f"Медіана (без бібліотек): {median_manual_value}")

# 3.2. Обчислення медіани з NumPy
median_numpy = np.median(sum_purchases)
print(f"Медіана (NumPy): {median_numpy}")

# Порівняння середнього і медіани
print("Порівняння вибіркового середнього і медіани:")
print(f"Вибіркове середнє: {mean_numpy}")
print(f"Медіана: {median_numpy}")

# Коментар
print("\nКоментар: Вибіркове середнє краще використовувати, коли дані розподілені рівномірно і немає значних викидів. Медіана краще підходить для даних з викидами, оскільки вона не так сильно на них впливає.")
```

  6.1. без використання жодних бібліотек, користуючись тільки Python.

Виведіть результат на екран. Проаналізуйте результат.

Виведіть додатково графік розсіювання викликавши код нижче і зробіть заключення, чи варто щось змінити при підрахунку кореляції, аби коефіцієнт Пірсона показував коректну оцінку лінійної залежності?

```python
visits = [ 6,  2, 10,  3,  9,  9,  2,  3,  6,  6, 10,  2,  5, 10,  3,  5,  4,
        7,  4,  2]
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Дані
sum_purchases = [316, 493, 367, 153, 224, 121, 322, 450, 196, 397, 281,
  239, 484, 124, 223, 448, 432, 255, 4310, 3524]
visits = [6, 2, 10, 3, 9, 9, 2, 3, 6, 6, 10, 2, 5, 10, 3, 5, 4, 7, 4, 2]

# Розрахунок коефіцієнта кореляції Пірсона без бібліотек
def pearson_corr(x, y):
    mean_x, mean_y = sum(x) / len(x), sum(y) / len(y)
    num = sum((xi - mean_x) * (yi - mean_y) for xi, yi in zip(x, y))
    den_x = sum((xi - mean_x) ** 2 for xi in x) ** 0.5
    den_y = sum((yi - mean_y) ** 2 for yi in y) ** 0.5
    return num / (den_x * den_y)

corr_manual = pearson_corr(sum_purchases, visits)
print(f"Коефіцієнт кореляції Пірсона (без бібліотек): {corr_manual}")

# Використання numpy
corr_numpy = np.corrcoef(sum_purchases, visits)[0, 1]
print(f"Коефіцієнт кореляції Пірсона (NumPy): {corr_numpy}")

# Побудова графіка розсіювання
plt.scatter(visits, sum_purchases)
plt.xlabel("Кількість відвідувань")
plt.ylabel("Сума покупок")
plt.title("Графік розсіювання: відвідування vs сума покупок")
plt.show()
```

**Завдання 7**. В попередньому завданні ми помітили, що викиди нам заважають отримати "правдиве" значення кореляції. Давайте вилучимо записи, які відповідають значенням з викидами в сумі покупок і обчисліть коеф. кореляції. Для цього поставте `threshold`, який дорінює середньому значенню + 2 стандартних вдіхилля суми покупок і всі значення, які більше за нього, будемо вважати викидами (є і інші правила для визначення викидів - з ним и познайомимось пізніше).

Вилучіть значення з обох масивів і обчисліть коеф. кореляції Пірсона з допомогою `numpy` для кількості візитів і суми покупок без записів-викидів. Виведіть результат на екран, побудуйте діаграму розсіювання і проаналізуйте результат: що можемо сказати про залежність в даних?

```python
import numpy as np
import matplotlib.pyplot as plt

# Дані
sum_purchases = np.array([316, 493, 367, 153, 224, 121, 322, 450, 196, 397,
                          281, 239, 484, 124, 223, 448, 432, 255, 4310, 3524])
visits = np.array([6, 2, 10, 3, 9, 9, 2, 3, 6, 6, 10, 2, 5, 10, 3, 5, 4, 7, 4, 2])

# Обчислення середнього та стандартного відхилення
mean_purchases = np.mean(sum_purchases)
std_purchases = np.std(sum_purchases)

# Визначення порогу для викидів (середнє + 2 стандартні відхилення)
threshold = mean_purchases + 2 * std_purchases

# Фільтрація даних без викидів
filtered_indices = sum_purchases <= threshold
filtered_purchases = sum_purchases[filtered_indices]
filtered_visits = visits[filtered_indices]

# Обчислення коефіцієнта кореляції Пірсона
correlation_filtered = np.corrcoef(filtered_purchases, filtered_visits)[0, 1]

# Візуалізація
plt.scatter(filtered_purchases, filtered_visits, color='blue', label="Без викидів")
plt.title('Scatter Plot of Purchases vs. Visits (Without Outliers)')
plt.xlabel('Sum of Purchases')
plt.ylabel('Number of Visits')
plt.legend()
plt.show()

# Вивід результату
correlation_filtered
```
