### Завдання: Візуалізація функцій експоненти, логарифму, і модифікованої синусоїди

**Мета:** Створити фігуру з трьома панелями: на першій показати графік експоненціальної функції, на другій - логарифмічну функцію, а на третій - модифіковану синусоїду.

1. Імпортуйте необхідні бібліотеки:


```python
import numpy as np
import matplotlib.pyplot as plt
```


2. Використовуючи фукнціонал `numpy` створіть масив `x` з 400 точок в діапазоні між -2 та 2 і масив `x_positive` в діапазоні між 0.1 і 2 також з 400 точок. Другий масив в нас буде для логарифмічної функції, перший з більш широким діапазоном - для інших функцій.


```python
# Створення масивів
x = np.linspace(-2, 2, 400)
x_positive = np.linspace(0.1, 2, 400)

x[:5], x_positive[:5]

```


3. Тут маємо кілька кроків.

  3.1. Ініціалізуйте фігуру Matplotlib з розміром, достатнім для трьох панелей (рекомендую спробувати експериментально виявити добрий розмір).   
  3.2. На першій панелі візуалізуйте експоненціальну функцію `y = e^x`. Використовуйте відповідну змінну `x`.  
  3.2. На другій панелі покажіть логарифмічну функцію `y = ln(x)`.  
  3.4. На третій панелі покажіть модифіковану синусоїду `y = sin(x) + x`.  

  Відобразіть графік.


```python
# 3.1: Ініціалізація фігури з трьома панелями
fig, axes = plt.subplots(3, 1, figsize=(8, 12))  # Розмір обрано експериментально

# 3.2: Експоненціальна функція y = e^x
y_exponential = np.exp(x)
axes[0].plot(x, y_exponential, label='y = e^x', color='blue')
axes[0].set_title('Експоненціальна функція')
axes[0].set_xlabel('x')
axes[0].set_ylabel('y')
axes[0].legend()
axes[0].grid(True)

# 3.3: Логарифмічна функція y = ln(x)
y_logarithmic = np.log(x_positive)
axes[1].plot(x_positive, y_logarithmic, label='y = ln(x)', color='green')
axes[1].set_title('Логарифмічна функція')
axes[1].set_xlabel('x')
axes[1].set_ylabel('y')
axes[1].legend()
axes[1].grid(True)

# 3.4: Модифікована синусоїда y = sin(x) + x
y_modified_sine = np.sin(x) + x
axes[2].plot(x, y_modified_sine, label='y = sin(x) + x', color='red')
axes[2].set_title('Модифікована синусоїда')
axes[2].set_xlabel('x')
axes[2].set_ylabel('y')
axes[2].legend()
axes[2].grid(True)

# Відображення графіка
plt.tight_layout()
plt.show()
```

4. Модифікація і довершення графіку.  
  4.1. Для кожної панелі встановіть назву графіка та підписи до осей х та у (можна так і назвати `x` та формула функції).  
  4.2. Використайте різні кольори для кожного графіка.  


```python
import matplotlib.pyplot as plt
import numpy as np

# Створення даних
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)

# Створення підграфіків
fig, axs = plt.subplots(2, 1, figsize=(8, 6))

# Перший графік (sin(x))
axs[0].plot(x, y1, color='blue')  # Використовуємо синій колір
axs[0].set_title('Графік функції y = sin(x)')
axs[0].set_xlabel('x')
axs[0].set_ylabel('sin(x)')

# Другий графік (cos(x))
axs[1].plot(x, y2, color='red')  # Використовуємо червоний колір
axs[1].set_title('Графік функції y = cos(x)')
axs[1].set_xlabel('x')
axs[1].set_ylabel('cos(x)')

# Відображення графіків
plt.tight_layout()  # Для зручного розміщення підписів
plt.show()
```

5. А тепер зобразіть ті самі 3 трафіки тих самих кольорів тільки на одному графіку. Визначіть назву кожної з ліній для легенди та обовʼязково відобразіть легенду.

```python
# Створення даних
x = np.linspace(0, 10, 100)
y1 = np.sin(x)
y2 = np.cos(x)
y3 = np.tan(x)

# Побудова графіків
plt.plot(x, y1, color='blue', label='y = sin(x)')  # Синій графік
plt.plot(x, y2, color='red', label='y = cos(x)')   # Червоний графік
plt.plot(x, y3, color='green', label='y = tan(x)') # Зелений графік

# Обмежуємо ось y, щоб уникнути величезних значень тангенса
plt.ylim(-10, 10)

# Встановлюємо назви осей
plt.xlabel('x')
plt.ylabel('y')

# Додаємо легенду
plt.legend()

# Додаємо заголовок
plt.title('Графіки функцій sin(x), cos(x) та tan(x)')

# Відображаємо графік
plt.show()
```
